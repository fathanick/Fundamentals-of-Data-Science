Data visualization is the process of turning data into visual forms such as charts, graphs, or maps. It helps us see patterns, trends, and relationships that are hard to notice in raw numbers.

In this tutorial, we will learn to create **7 types of visualizations** in Python using the `matplotlib` and `seaborn` libraries, applied to a real-world supermarket sales dataset.

**Chart types we will cover:**
1. Tables
2. Bar Chart
3. Histogram
4. Line Chart
5. Scatterplot
6. Pie Chart
7. Dual-Axis Chart
8. Heatmap *(Bonus)*

```python
# Install libraries (if needed)
# !pip install matplotlib seaborn pandas

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set plotting style
sns.set_theme(style="whitegrid")
```

## Dataset: Supermarket Sales

We will use the **Supermarket Sales** dataset from Kaggle. It contains **1,000 transactions** from a supermarket chain with **17 columns** including:

- `Branch` — store branch (A, B, or C)
- `Product line` — category of product sold
- `Gender` — customer gender
- `Payment` — payment method used
- `Total` — total sale amount (USD)
- `Rating` — customer satisfaction rating (1–10)
- `Date` / `Time` — when the transaction occurred

This rich dataset is perfect for exploring different visualization techniques.

```python
url = "https://raw.githubusercontent.com/fathanick/Fundamentals-of-Data-Science/main/Data%20Visualization/supermarket_sales.csv"
df = pd.read_csv(url)
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

```python
df.describe()
```

---
## 1. Tables as Visualization

Tables are useful when working with a **small number of data points**. They present data in a clear and structured way, showing exact values.

Let's create a simple summary table showing the average customer rating per product line.

```python
summary = df.groupby('Product line')['Rating'].mean().round(2).reset_index()
summary.columns = ['Product Line', 'Average Rating']
summary.sort_values('Average Rating', ascending=False)
```

> **Note**: Tables work well for looking up specific numbers, but they are **not ideal for seeing patterns or trends**. That is where charts come in.

---
## 2. Bar Chart

Bar charts are perfect for **comparing categorical data**. Best practices:
- Order bars by value for nominal (unordered) categories.
- Use horizontal bars when category labels are long.

### 2a. Total Sales by Product Line (Horizontal Bar Chart)

```python
sales_by_product = df.groupby('Product line')['Total'].sum().sort_values(ascending=True)

plt.figure(figsize=(8, 5))
sales_by_product.plot(kind='barh', color='steelblue')
plt.xlabel('Total Sales (USD)')
plt.ylabel('Product Line')
plt.title('Total Sales by Product Line')
plt.tight_layout()
plt.show()
```

### 2b. Grouped Bar Chart — Sales by Product Line and Gender

We can add another dimension by grouping bars, allowing us to compare across two categorical variables simultaneously.

```python
sales_gender = df.groupby(['Product line', 'Gender'])['Total'].sum().unstack()

sales_gender.plot(kind='bar', figsize=(10, 5), color=['#E91E63', '#2196F3'])
plt.title('Total Sales by Product Line and Gender')
plt.ylabel('Total Sales (USD)')
plt.xlabel('Product Line')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

---
## 3. Histogram

A histogram is used to **summarize the distribution** of a numerical data sample:
- The **x-axis** represents bins (intervals of values).
- The **y-axis** represents frequency (how many data points fall in each bin).

Histograms help answer: *Is the data skewed? Are there outliers? Is the distribution normal?*

```python
plt.figure(figsize=(8, 5))
plt.hist(df['Rating'], bins=10, color='orange', edgecolor='black', alpha=0.7)
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Customer Ratings')
plt.tight_layout()
plt.show()
```

---
## 4. Line Chart

Line charts are suited for showing **temporal evolutions** — data collected over time. Key rules:
- Use for **continuous data** (time series).
- Avoid using line charts for categorical comparisons.

Let's plot daily total sales over the 3-month period.

```python
daily_sales = df.groupby('Date')['Total'].sum().reset_index()

plt.figure(figsize=(12, 5))
plt.plot(daily_sales['Date'], daily_sales['Total'], marker='o', markersize=3, linewidth=1, color='teal')
plt.xlabel('Date')
plt.ylabel('Total Sales (USD)')
plt.title('Daily Total Sales Over Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

---
## 5. Scatterplot

Scatterplots show the **relationship between two different numerical variables**:
- Each point represents one observation.
- Helps reveal correlation, clusters, and outliers.
- Adding color (`hue`) introduces a third variable.

Let's explore how unit price relates to total sale amount.

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Unit price', y='Total', hue='Product line', alpha=0.7)
plt.title('Unit Price vs Total Sale Amount')
plt.xlabel('Unit Price (USD)')
plt.ylabel('Total (USD)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
plt.tight_layout()
plt.show()
```

---
## 6. Pie Chart

Pie charts are used to summarize **categorical data as parts of a whole**:
- Best for showing proportions or percentages.
- Slice values should add up to 100%.
- Order slices by size for readability.
- Avoid using pie charts when there are many categories (use bar charts instead).

Let's show the distribution of payment methods used by customers.

```python
payment_counts = df['Payment'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%',
        startangle=140, colors=['#FF9800', '#4CAF50', '#2196F3'])
plt.title('Payment Method Distribution')
plt.tight_layout()
plt.show()
```

---
## 7. Dual-Axis Chart

A dual-axis chart combines **two different chart types** (e.g., bar and line) on **two y-axes**. It is useful when you want to:
- Compare two related but different-scale metrics side by side.
- Show both volume and quality at the same time.

Let's compare average rating (bar) vs total sales (line) per branch.

```python
branch_stats = df.groupby('Branch').agg(
    avg_rating=('Rating', 'mean'),
    total_sales=('Total', 'sum')
).reset_index()

fig, ax1 = plt.subplots(figsize=(8, 5))

ax1.bar(branch_stats['Branch'], branch_stats['avg_rating'], color='steelblue', alpha=0.7, label='Avg Rating')
ax1.set_xlabel('Branch')
ax1.set_ylabel('Average Rating', color='steelblue')
ax1.tick_params(axis='y', labelcolor='steelblue')

ax2 = ax1.twinx()
ax2.plot(branch_stats['Branch'], branch_stats['total_sales'], color='red', marker='o', linewidth=2, label='Total Sales')
ax2.set_ylabel('Total Sales (USD)', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.title('Average Rating vs Total Sales by Branch')
fig.tight_layout()
plt.show()
```

---
## 8. Heatmap *(Bonus)*

Heatmaps are great for showing **correlations or patterns across two categorical dimensions** using color intensity:
- Darker / warmer colors = higher values.
- Great for time-based patterns (e.g., day vs hour).

Let's visualize average sales by product line and day of the week.

```python
df['DayOfWeek'] = df['Date'].dt.day_name()
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

heatmap_data = df.pivot_table(values='Total', index='Product line', columns='DayOfWeek', aggfunc='mean')
heatmap_data = heatmap_data[day_order]

plt.figure(figsize=(12, 5))
sns.heatmap(heatmap_data, annot=True, fmt='.0f', cmap='YlOrRd', linewidths=0.5)
plt.title('Average Sales by Product Line and Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Product Line')
plt.tight_layout()
plt.show()
```

---
## Choosing the Right Chart — Summary

| Goal | Recommended Chart |
|---|---|
| Compare categories | Bar chart |
| Show distribution | Histogram |
| Show trends over time | Line chart |
| Show relationships | Scatterplot |
| Show proportions | Pie chart |
| Compare two metrics | Dual-axis chart |
| Show patterns across 2 dimensions | Heatmap |

> **Key takeaway**: Choosing the right chart type depends on the **type of data** you have and the **question** you are trying to answer.
