## 2. Python Libraries for Data Science

We will use three core libraries throughout this course:
- **NumPy** — fast numerical computing and random data generation
- **Pandas** — working with tabular data (like spreadsheets)
- **Matplotlib** — creating charts and visualizations

### 2.1 NumPy — Numerical Computing

NumPy (Numerical Python) is the foundation of scientific computing in Python.
Its main object is the **ndarray** (n-dimensional array), which supports
fast element-wise operations that would be much slower with plain Python lists.

```python
# Creating a NumPy array and computing basic statistics
scores = np.array([72, 85, 90, 68, 95, 78, 88])
print('Scores :', scores)
print('Mean   :', scores.mean())
print('Max    :', scores.max())
print('Sum    :', scores.sum())
```

NumPy can also **generate random data**, which is useful for simulations
and testing. `np.random.randint` produces random integers, while
`np.random.normal` draws samples from a normal (bell-curve) distribution.

```python
# Generating random data with NumPy
np.random.seed(42)  # set seed so results are reproducible

rand_ints   = np.random.randint(1, 101, size=10)   # 10 integers in [1, 100]
rand_floats = np.random.normal(loc=70, scale=10, size=10)  # normal dist

print('Random integers:', rand_ints)
print('Normal samples :', np.round(rand_floats, 2))
```

### 2.2 Pandas — Data Handling

Pandas provides two key data structures:
- **Series** — a one-dimensional labelled array (one column of data)
- **DataFrame** — a two-dimensional table, like a spreadsheet

Let's create a small DataFrame from a Python dictionary.

```python
# Creating a DataFrame from a dictionary
data = {
    'Name'         : ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Math_Score'   : [85, 72, 90, 68, 95],
    'Science_Score': [78, 88, 82, 74, 91],
}
df = pd.DataFrame(data)
df  # display the table
```

Two quick methods help you understand any DataFrame at a glance.
`.shape` tells you the number of rows and columns, while `.describe()`
returns summary statistics (count, mean, std, min, quartiles, max)
for every numeric column.

```python
# Exploring the DataFrame
print('Shape:', df.shape)  # (rows, columns)
print()
print(df.describe())       # summary statistics for numeric columns
```

### 2.3 Matplotlib — Basic Visualization

Matplotlib is the most widely used plotting library in Python. We import
its `pyplot` module as `plt`. A typical workflow is:
1. Call `plt.figure()` to set the canvas size.
2. Call a plot function (`plt.bar`, `plt.hist`, `plt.scatter`, …).
3. Add labels and a title, then call `plt.show()`.

```python
# Bar chart — comparing Math Scores across students
plt.figure(figsize=(6, 4))
plt.bar(df['Name'], df['Math_Score'], color='steelblue')
plt.title('Math Scores by Student')
plt.xlabel('Student')
plt.ylabel('Score')
plt.tight_layout()
plt.show()
```

A **histogram** groups values into bins and shows how often each range
appears. It is the go-to chart for understanding the *shape* of a
distribution — whether it is symmetric, skewed, or has multiple peaks.

```python
# Histogram — distribution of Science Scores
plt.figure(figsize=(6, 4))
plt.hist(df['Science_Score'], bins=5, color='coral', edgecolor='black')
plt.title('Distribution of Science Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

## 4. Basic Data Visualization

Good visualizations make patterns in data immediately visible.
In this section we create two complementary plots for the Tips dataset
and practice reading what they tell us.

### 4.1 Histogram with Mean and Median Lines

We overlay vertical lines for the **mean** (red dashed) and **median**
(orange dashed). If the mean is to the right of the median, the distribution
is **right-skewed** — a few large values pull the average up.

```python
# Histogram of total_bill with mean and median lines
plt.figure(figsize=(8, 4))
plt.hist(tips['total_bill'], bins=20, color='steelblue',
         edgecolor='black', alpha=0.7)
plt.axvline(tips['total_bill'].mean(),
            color='red', linestyle='--',
            label=f"Mean = {tips['total_bill'].mean():.2f}")
plt.axvline(tips['total_bill'].median(),
            color='orange', linestyle='--',
            label=f"Median = {tips['total_bill'].median():.2f}")
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill ($)')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.show()
```

### 4.2 Scatter Plot — Relationship Between Two Variables

A **scatter plot** shows one dot per observation. Its x-position is one
variable and its y-position is another. An upward trend (dots sloping
from lower-left to upper-right) indicates a **positive relationship**.

```python
# Scatter plot: total_bill vs tip
plt.figure(figsize=(8, 5))
plt.scatter(tips['total_bill'], tips['tip'],
            alpha=0.6, color='coral',
            edgecolors='black', linewidth=0.5)
plt.title('Total Bill vs Tip Amount')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.tight_layout()
plt.show()
```

### 4.3 Interpretation

From these two plots we can observe:

1. **Total bill is right-skewed** — the mean ($19.79) is higher than the
   median ($17.80), meaning a few large bills pull the average up. Most
   meals cost between \$10 and \$25.

2. **Tips increase with the bill** — the scatter plot shows a clear upward
   trend: the bigger the bill, the larger the tip tends to be.
