## 2. Introduction to Data Science Libraries

We will use three core libraries throughout this tutorial:
- **NumPy** — for numerical computing and random data generation
- **Pandas** — for working with tabular data
- **Matplotlib** — for creating visualizations

### 2.1 NumPy — Numerical Computing

NumPy (Numerical Python) is the foundation of scientific computing in Python.
Its main object is the **array**, which is like a Python list but much faster for
mathematical operations.

```python
# Creating arrays
a = np.array([1, 2, 3, 4, 5])
print('Array:', a)
print('Mean :', a.mean())
print('Sum  :', a.sum())
```

NumPy also makes it easy to **generate random data** — very useful for simulations
and testing. We can create random integers with `np.random.randint` or draw samples
from a normal distribution with `np.random.normal`.

```python
# Generating random data
random_integers = np.random.randint(1, 100, size=10)  # 10 random integers between 1 and 99
print('Random integers:', random_integers)

random_floats = np.random.normal(loc=70, scale=10, size=10)  # 10 values from a normal distribution
print('Random floats  :', np.round(random_floats, 2))
```

### 2.2 Pandas — Data Handling

Pandas provides two main data structures:
- **Series** — a one-dimensional labelled array (like a single column)
- **DataFrame** — a two-dimensional table (like a spreadsheet)

Let's create a simple DataFrame and explore it.

```python
# Creating a DataFrame from a dictionary
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Math_Score': [85, 72, 90, 68, 95],
    'Science_Score': [78, 88, 82, 74, 91]
}
df = pd.DataFrame(data)
df
```

Two quick methods help you understand any DataFrame at a glance. `.shape` tells you

### 2.3 Matplotlib — Basic Visualization

Matplotlib is the most widely used plotting library in Python. We will use its
`pyplot` module (imported as `plt`) to create charts.

```python
# Simple bar chart of math scores
plt.figure(figsize=(6, 4))
plt.bar(df['Name'], df['Math_Score'], color='steelblue')
plt.title('Math Scores')
plt.xlabel('Student')
plt.ylabel('Score')
plt.tight_layout()
plt.show()
```

A **histogram** groups values into bins and shows how often each range appears.

## 3. Random Variables and Basic Statistics

In this section we explore what random variables are and learn how to summarize
data using basic statistics.

### 3.1 Discrete vs Continuous Random Variables

A **random variable** is a variable whose value is determined by the outcome of a
random process.

| Type | Description | Examples |
|------|-------------|----------|
| **Discrete** | Takes countable, distinct values | Dice roll (1–6), number of students in a class |
| **Continuous** | Takes any value in a range | Height (170.5 cm), temperature (36.7 °C) |

**Think of it this way:** if you can *count* the possible values, it is discrete.
If you need to *measure* them, it is continuous.

### 3.2 Generating Random Data with NumPy

Let's generate examples of both types.

```python
# Set a random seed for reproducibility
np.random.seed(42)

# Discrete: simulating 100 dice rolls (values 1-6)
dice_rolls = np.random.randint(1, 7, size=100)
print('First 20 dice rolls:', dice_rolls[:20])

# Continuous: simulating 100 height measurements (mean=165 cm, std=8 cm)
heights = np.random.normal(loc=165, scale=8, size=100)
print('First 10 heights (cm):', np.round(heights[:10], 1))
```

Let's visualize both distributions side by side.

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Discrete distribution
axes[0].hist(dice_rolls, bins=range(1, 8), color='mediumpurple',
             edgecolor='black', align='left', rwidth=0.8)
axes[0].set_title('Discrete: Dice Rolls')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Frequency')
axes[0].set_xticks(range(1, 7))

# Continuous distribution
axes[1].hist(heights, bins=15, color='salmon', edgecolor='black')
axes[1].set_title('Continuous: Heights (cm)')
axes[1].set_xlabel('Height (cm)')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

### 3.3 Basic Statistics

Statistics help us **summarize** a dataset with just a few numbers. The three most
fundamental measures are:

| Statistic | What it tells you | Analogy |
|-----------|-------------------|---------|
| **Mean** | The average value — the "center of gravity" of the data | If you balanced all values on a seesaw, the mean is the balance point |
| **Median** | The middle value when data is sorted — resistant to extreme values | The person standing in the exact middle of a line |
| **Standard Deviation (std)** | How spread out the values are around the mean | A small std means values are clustered; a large std means they are scattered |

Let's compute these for our generated data.

```python
# Statistics for dice rolls (discrete)
print('=== Dice Rolls ===')
print(f'Mean  : {np.mean(dice_rolls):.2f}')
print(f'Median: {np.median(dice_rolls):.2f}')
print(f'Std   : {np.std(dice_rolls):.2f}')

print()

# Statistics for heights (continuous)
print('=== Heights (cm) ===')
print(f'Mean  : {np.mean(heights):.2f}')
print(f'Median: {np.median(heights):.2f}')
print(f'Std   : {np.std(heights):.2f}')
```

**Interpretation:**
- The dice roll mean should be close to 3.5 (the theoretical average of a fair die).
- The heights mean should be close to 165 cm (the value we set when generating the data).
- The standard deviation of heights (~8) tells us most heights fall within roughly
  8 cm above or below the mean.

## 4. Case Study: Exploring the Tips Dataset

Now let's apply everything we've learned to a real dataset. We will use the
**Tips** dataset from the Seaborn library, which contains information about
restaurant bills and tips.

### 4.1 Loading the Data

```python
# Load the tips dataset from seaborn
tips = sns.load_dataset('tips')
print(f'Dataset shape: {tips.shape[0]} rows x {tips.shape[1]} columns')
tips.head()
```

### 4.2 Exploring the Data

Before computing anything, we should understand the structure of our data.

`tips.info()` shows the column names, data types, and how many non-null values
each column has. It is the first thing to check to spot missing data or unexpected types.

```python
# Data types and non-null counts
tips.info()
```

`tips.describe()` computes summary statistics (count, mean, std, min, quartiles, max)
for all numeric columns at once — a quick way to get a feel for the data's scale and spread.

```python
# Summary statistics for numeric columns
tips.describe()
```

### 4.3 Computing Statistics

Let's focus on two numeric columns: **total_bill** and **tip**.

```python
for col in ['total_bill', 'tip']:
    print(f'--- {col} ---')
    print(f'  Mean  : {tips[col].mean():.2f}')
    print(f'  Median: {tips[col].median():.2f}')
    print(f'  Std   : {tips[col].std():.2f}')
    print()
```

### 4.4 Visualization

Let's create two visualizations to better understand the data.

```python
# Histogram of total bills
plt.figure(figsize=(8, 4))
plt.hist(tips['total_bill'], bins=20, color='steelblue', edgecolor='black', alpha=0.7)
plt.axvline(tips['total_bill'].mean(), color='red', linestyle='--', label=f"Mean = {tips['total_bill'].mean():.2f}")
plt.axvline(tips['total_bill'].median(), color='orange', linestyle='--', label=f"Median = {tips['total_bill'].median():.2f}")
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill ($)')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.show()
```

A **scatter plot** is ideal for exploring the *relationship* between two numeric variables.
Each dot represents one restaurant visit; its x-position is the bill total and
its y-position is the tip amount.

```python
# Scatter plot: total bill vs tip
plt.figure(figsize=(8, 5))
plt.scatter(tips['total_bill'], tips['tip'], alpha=0.6, color='coral', edgecolors='black', linewidth=0.5)
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.tight_layout()
plt.show()
```

### 4.5 Interpretation

From our analysis of the Tips dataset, we can observe:

1. **Total bill distribution is right-skewed** — the mean is higher than the median,
   indicating that a few large bills pull the average up. Most bills are between \$10 and \$25.

2. **Tips also show a right-skewed pattern** — most tips are modest (around \$2–\$4),
   but some customers leave substantially larger tips.

3. **There is a positive relationship** between total bill and tip amount — as the
   bill increases, the tip tends to increase as well, which is intuitive.
